```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import models, layers
```

### ANN Architecture Explanation

- Input layer represents scaled numerical features
- Hidden layers capture non-linear relationships
- ReLU activation helps learn complex patterns
- Sigmoid activation is used for binary classification


```python
def create_model(hp=None):

    # --------- HYPERPARAMETER TUNING MODEL ---------
    if hp is not None:
        model = models.Sequential()

        # First hidden layer
        model.add(layers.Dense(
            units=hp.Choice('units_1', values=[32, 64, 128]),
            activation='relu',
            input_shape=(x_train_scaled.shape[1],)
        ))

        # Optional second hidden layer
        if hp.Boolean('second_layer'):
            model.add(layers.Dense(
                units=hp.Choice('units_2', values=[16, 32]),
                activation='relu'
            ))

        # Output layer
        model.add(layers.Dense(1, activation='sigmoid'))

        # Compile
        model.compile(
            optimizer='adam',
            loss='binary_crossentropy',
            metrics=['accuracy']
        )

    # --------- NORMAL MODEL (NO TUNING) ---------
    else:
        model = models.Sequential([
            layers.Dense(64, activation='relu', input_shape=(x_train_scaled.shape[1],)),
            layers.Dense(32, activation='relu'),
            layers.Dense(16, activation='relu'),
            layers.Dense(8, activation='relu'),
            layers.Dense(1, activation='sigmoid')
        ])

        model.compile(
            optimizer='adam',
            loss='binary_crossentropy',
            metrics=['accuracy']
        )

    return model
```

```python
initial_model = create_model()
initial_model.summary()
```

```python
history = initial_model.fit(
    x_train_scaled , y_train ,
    epochs = 50 ,
    batch_size = 32 ,
    validation_split = 0.20
)
```

```python
loss , accuracy = initial_model.evaluate(x_test_scaled , y_test)
print(f'loss of initial model: {loss}')
print(f'accuuracy of initial model: {accuracy}')
```

Random Forest was used along with Recursive Feature Elimination (RFE) to identify the most important features.
These selected features were then used to train the ANN model to reduce noise and improve generalization.


```python
#best option is to check hyperparamter tuning
!pip install keras-tuner --upgrade
```

```python
import keras_tuner as kt

tuner = kt.RandomSearch(
    create_model,
    objective = 'val_accuracy',
    max_trials = 5,
    directory = 'keras_tuning',
    project_name = 'Credit Card Default Prediction'
)
```

```python
import tensorflow as tf
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss' , patience=20)
```

```python
# “You are allowing the model to automatically explore different configurations and select the one that performs best.”

tuner.search(
    x_train_scaled , y_train,
    epochs = 25,
    validation_split = 0.20,
    callbacks = [stop_early]
)
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials = 1)[0]
print(best_hps)

for i in best_hps.values:
  print(f"{i} : {best_hps.get(i)}")
```

```python
best_model = tuner.hypermodel.build(best_hps)
best_history = best_model.fit(
    x_train_scaled, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.20
)
```

```python
plt.figure(figsize=(8,3))
plt.plot(best_history.history['loss'],label='training loss')
plt.plot(best_history.history['val_loss'],label='validation loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()


plt.figure(figsize=(8,3))
plt.plot(best_history.history['accuracy'],label='training accuracy')
plt.plot(best_history.history['val_accuracy'],label='val_accuracy')
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
```

```python
best_loss, best_accuracy = best_model.evaluate(x_test_scaled, y_test)
print(f"Best model Loss : {best_loss}")
print(f"Best model Accuracy : {best_accuracy}")
```
